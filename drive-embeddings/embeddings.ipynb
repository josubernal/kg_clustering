```python
from pykeen.triples import TriplesFactory
from pykeen.pipeline import pipeline

print("This will take about 50 minutes. If you want to drink coffee or say hello to a friend, now is your opportunity. Take a small break. :)")
# Load triples and train the model
tf = TriplesFactory.from_path('/content/gdrive/MyDrive/drive-embeddings/data/triples.tsv')
training, testing = tf.split()

device = "cuda" if torch.cuda.is_available() else "cpu"

result = pipeline(
    training=training,
    testing=testing,
    model="TransE",
    model_kwargs=dict(embedding_dim=64),
    training_kwargs=dict(num_epochs=5),
    optimizer_kwargs=dict(lr=0.01),
    negative_sampler_kwargs=dict(num_negs_per_pos=1),
    random_seed=2025,
    device=device
)
```

```python
# STORE

model_name = "./gdrive/MyDrive/drive-embeddings/kg"
result.save_to_directory(model_name)

```
